```python
features_train_ohe = features_train.select_dtypes(include='object').columns.tolist()
display(features_train_ohe)

num_features = features_train.select_dtypes(exclude='object').columns.tolist()
num_features
```

```python

encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)

encoder_ohe.fit(features_train[features_train_ohe])

features_train[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_train[features_train_ohe])

features_train = features_train.drop(features_train_ohe, axis=1)

scaler = StandardScaler()

features_train[num_features] = scaler.fit_transform(features_train[num_features])

features_train.head()
```

```python
features_test[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_test[features_train_ohe])

features_test = features_test.drop(features_train_ohe, axis=1)

#—Å–∫–µ–ª–µ—Ä–æ–º, –∫–æ—Ç–æ—Ä—ã–π –æ–±—É—á–µ–Ω –Ω–∞ –¢–†–ï–ù–ò–†–û–í–û–ß–ù–û–ô –í–´–ë–û–†–ö–ï, –º–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º —Ç–µ—Å—Ç–æ–≤—É—é
features_test[num_features] = scaler.transform(
    features_test[num_features]
)

#—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
features_test.head()
```

```python

param_grid = {
    "bootstrap": [True],
    "max_depth": [20, 30, 40],
    "max_features": [2, 3, 5],
    "min_samples_leaf": [2, 3],
    "min_samples_split": [2, 3],
    "n_estimators": [20, 40],
}

model_rfc = RandomForestClassifier()

grid_search_rfc = GridSearchCV(
    estimator=model_rfc,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=3,
    n_jobs=-1,
    verbose=2,
)
grid_search_rfc.fit(features_train, target_train)

grid_search_rfc.best_score_
```

```python
catboost = CatBoostClassifier(random_seed=RANDOM_STATE, verbose=2)

catboost_params = {
    'iterations': [100, 200],
    'learning_rate': [0.05, 0.1, 0.001, 0.0001],
    'depth': [4, 6, 8],
    'l2_leaf_reg': [1, 3, 5],
    'border_count': [32, 64],
    'eval_metric': ['AUC'],
    'early_stopping_rounds': [10],
}



grid_search_catboost = GridSearchCV(catboost,
                                   catboost_params, 
                                   verbose=3,
                                   cv=4,
                                   n_jobs=-1,
                                   scoring='roc_auc')


grid_search_catboost.fit(features_train, target_train)
print(grid_search_catboost.best_score_)
print(grid_search_catboost.best_params_)
```

```python
param_grid = {
    "learning_rate": [0.1],
    "max_depth": [1,2,3],
}
LGBMC = lgb.LGBMClassifier(verbose = 2, force_row_wise=True)
grid_search_lgbm = GridSearchCV(
    estimator=LGBMC,
    param_grid=param_grid,
    cv=2,
    n_jobs=-1,
    verbose=0,
    scoring='roc_auc'
    )
grid_search_lgbm.fit(features_train, target_train)
grid_search_lgbm.best_score_
```
